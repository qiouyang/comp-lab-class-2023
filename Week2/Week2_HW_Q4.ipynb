```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import os
```

```python
# Read the EDR file
path = './Data/1hz3_T310.run.25000000.energy.xvg'
data = np.loadtxt(path,comments=['@', '#'],unpack=True)
```

```python
data = data.T
```

```python
# @ s0 legend "Potential"
# @ s1 legend "Kinetic En."
# @ s2 legend "Total Energy"
# @ s3 legend "Temperature"
# @ s4 legend "Pressure"
# @ s5 legend "Volume"

# plot Potential vs. Time -> x = data[:, 0], y = data[:, 1]
```

```python
import numpy as np
import matplotlib.pyplot as plt

labels = [
    "Potential",
    "Kinetic En.",
    "Total Energy",
    "Temperature",
    "Pressure",
    "Volume"
]

for j in range(1, 7):
    x = data[:, 0]
    y = data[:, j]
    plt.plot(x, y)
    plt.xlabel('Time') 
    plt.ylabel(labels[j-1])
    plt.title(f'{labels[j-1]} vs. Time')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

time = data[:, 0]
kinetic_energy = data[:, 2]
temperature = data[:, 4]

fig, ax1 = plt.subplots(figsize=(12,4))
ax1.plot(time, temperature, c='r')
ax2 = ax1.twinx()
ax2.plot(time, kinetic_energy, c='b')

ax1.set_ylabel('Temperature')
ax2.set_ylabel('Kinetic En.')
plt.title('Kinetic En. & Temperature vs. Time')

ax1.set_ylim(300, 322)
ax2.set_ylim(34250, 37000)

plt.show()

fig.savefig('test.png')
```
