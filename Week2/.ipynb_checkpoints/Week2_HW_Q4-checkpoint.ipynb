### a) Reads in the data file Data/1hz3_T310.run.25000000.energy.xvg

```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import os
```

```python
# Read the xvg file
path = './Data/1hz3_T310.run.25000000.energy.xvg'
data = np.loadtxt(path,comments=['@', '#'],unpack=True)
```

```python
# Check the shape of the data (x = 7-1/quantity, y = 501/time)
data.shape
```

```python
# reverse the data
data = data.T
```

```python
# comfirm the data shape (x = 501/time, y = 7-1/quantity)
data.shape
```

### b) Make 6 separate plots for each quantity vs time, with time in ns

```python
# @ s0 legend "Potential"
# @ s1 legend "Kinetic En."
# @ s2 legend "Total Energy"
# @ s3 legend "Temperature"
# @ s4 legend "Pressure"
# @ s5 legend "Volume"

# plot Potential vs. Time -> x = data[:, 0], y = data[:, 1]
```

```python
time_in_ns = data[:,0]*0.01

labels = [
    "Potential",
    "Kinetic En.",
    "Total Energy",
    "Temperature",
    "Pressure",
    "Volume"
]

for j in range(1, 7):
    x = time_in_ns
    y = data[:, j]
    plt.plot(x, y)
    plt.xlabel('Time (ns)') 
    plt.ylabel(labels[j-1])
    plt.title(f'{labels[j-1]} vs. Time')
    plt.show()
```

### c) Make a plot that has kinetic energy and temperature on the same plot, where the left axis is temperature and the right axis is kinetic energy, using different colors and different axis labels

```python
import numpy as np
import matplotlib.pyplot as plt

time = data[:, 0]
kinetic_energy = data[:, 2]
temperature = data[:, 4]

fig, ax1 = plt.subplots(figsize=(12,4))
ax1.plot(time, temperature, c='r')
ax2 = ax1.twinx()
ax2.plot(time, kinetic_energy, c='b')

ax1.set_ylabel('Temperature')
ax2.set_ylabel('Kinetic En.')
plt.title('Kinetic En. & Temperature vs. Time')

ax1.set_ylim(300, 322)
ax2.set_ylim(34250, 37000)

plt.show()

fig.savefig('test.png')
```
