```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load data using np.loadtxt
rmsd = np.loadtxt('rmsd_jz4.xvg', comments=['@', '#'])

# Slice columns into time and potential  
x_time = rmsd[:,0]
y_rmsd = rmsd[:,1]

# Creating the plot
plt.figure(figsize=(10, 6))
plt.plot(x_time, y_rmsd, marker='o', color='b')

# Setting title and labels
plt.title("RMSD of JZ4_Heavy after lsq fit to Backbone")
plt.xlabel("Time (ns)")
plt.ylabel("RMSD (nm)")
plt.grid(True)

# Show plot
plt.savefig(f'../Figures/rmsd_jz4.png')
plt.show()
```

```python
# Load the data from the text file
data = np.loadtxt('dist.xvg', comments=['@', '#'])

# Separate the data into time and distance
times = data[:, 0]  # First column: Time
distances = data[:, 1]  # Second column: Distance

# Creating the plot
plt.figure(figsize=(10, 6))
plt.plot(times, distances, marker='o')

# Adding the title and labels
plt.title("Distance between JZ4 and Gln102")
plt.xlabel("Time (ps)")
plt.ylabel("Distance (nm)")

# Show plot
plt.savefig(f'../Figures/jz4_gln201_distance.png')
plt.show()
```

```python
# Load the data from the text file
data = np.loadtxt('angle.xvg', comments=['@', '#'])

# Splitting the data into two arrays: time_ps and angles_degrees
time_ps = data[:, 0]  # First column (Time in picoseconds)
angles_degrees = data[:, 1]  # Second column (Angles in degrees)

# Creating the plot
plt.figure(figsize=(10, 6))
plt.plot(time_ps, angles_degrees, marker='o', color='b')
plt.title("Average Angle: JZ4 & OAB_H12_Protein & r_102 & OE1")
plt.xlabel("Time (ps)")
plt.ylabel("Angle (degrees)")
plt.grid(True)
plt.xticks(range(0, 160, 10))  # Setting x-axis ticks for better readability
plt.tight_layout()

# Show plot
plt.savefig(f'../Figures/average_angles.png')
plt.show()
```
