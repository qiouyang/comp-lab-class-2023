```python
import matplotlib.pyplot as plt  
import numpy as np

# Load data using np.loadtxt
potential = np.loadtxt('1aki_potential.xvg', comments=['@', '#'])

# Slice columns into time and potential  
x_potential = potential[:,0]
y_potential = potential[:,1]

# Plot potential vs time
plt.plot(x_potential, y_potential)

# Label axes
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')

# Add title and show plot
plt.title('Energy Minimization') 

plt.savefig(f'../Figures/potential_jupyter.png')
plt.show()
```

```python
# Load pressure data
time_points, pressure_values = np.loadtxt('1aki_pressure.xvg', unpack=True, comments=['@','#'])

# Calculate 10ps running average
window_size = 10 # 10ps window
averaged_pressure = np.convolve(pressure_values, np.ones(window_size)/window_size, mode='valid') 

# Create figure and axes
fig, ax = plt.subplots()

# Plot raw pressure data
ax.plot(time_points, pressure_values, label='Raw data')  

# Plot running average
ax.plot(time_points[window_size//2:-window_size//2+1], averaged_pressure, label='10-ps Running Avg')

# Add labels and title
ax.set_xlabel('Time (ps)')
ax.set_ylabel('Pressure (bar)')
ax.set_title('Pressure')

# Add legend
ax.legend()

plt.savefig(f'../Figures/pressure_jupyter.png')
```
