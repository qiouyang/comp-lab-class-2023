### a) Read in the md trajectory of beta amyloid

```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load the beta amyloid PDB file
traj = md.load_xtc('./Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc',
                   top='./Data/1hz3_T310.start.nowater.gro')
```

```python
traj.xyz.shape
```

### b) Compute the end-end distance at each time, and the radius of gyration at each time

```python
def end_end(traj):
    """Compute the end to end distance of a trajectory"""
    # get the positions of the first and last atoms
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atoms
    dist = np.linalg.norm(first - last, axis=1)
    return dist

# compute the end to end distance
dist = end_end(traj)
print(dist)
```

```python
# radius of gyration
gyration = md.compute_rg(traj)
print(gyration)
```

### c) Plot the end-end distance and radius of gyration vs time on the same plot, clearly labeled (same y axis is fine this time)

```python
traj.time
```

```python
import numpy as np
import matplotlib.pyplot as plt

time = traj.time

fig, ax1 = plt.subplots(figsize=(12,4))
ax1.plot(time, dist, c='r')
ax2 = ax1.twinx()
ax2.plot(time, gyration, c='b')

ax1.set_ylabel('End-End Distance')
ax2.set_ylabel('Radius of Gyration')
plt.title('End-End Distance (Red) & Radius of Gyration (Blue) vs. Time')

plt.show()
```
