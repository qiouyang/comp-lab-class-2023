```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import os
```

```python
import pyedr

# Read the EDR file
path = './Data/1hz3_T310.run.25000000.edr'
dic = pyedr.edr_to_dict(path)

# The `verbose` optional parameter can be set to True to display the
# progress on stderr
dic = pyedr.edr_to_dict(path, verbose=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(dic['Time'], dic['Kinetic En.'])
plt.xlabel('Time') 
plt.ylabel('Kinetic En.')
plt.title('Kinetic En. vs. Time')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

time = dic['Time']
kinetic_energy = dic['Kinetic En.']
temperature = dic['Temperature']

fig, ax1 = plt.subplots(figsize=(12,4))
ax1.plot(time, temperature, c='r')
ax2 = ax1.twinx()
ax2.plot(time, kinetic_energy, c='b')

ax1.set_ylabel('Kinetic En.')
ax2.set_ylabel('Temperature')
plt.title('Kinetic En. & Temperature vs. Time')

plt.show()

fig.savefig('test.png')
```
